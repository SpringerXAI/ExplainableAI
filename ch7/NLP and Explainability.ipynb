```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import nltk

nltk.download('punkt')
nltk.download('wordnet')
```

```python
pd.set_option('display.max_colwidth', None)
covid_train = pd.read_csv('Covid_Train.csv')
covid_validation = pd.read_csv('Covid_Validation.csv')
covid_test = pd.read_csv('Covid_Test.csv')
```

```python
covid_train.head(3)
```

```python
covid_test.head(10)
```

```python
def replace_sentence(row):
  new_str = row['Text'].replace("[SEP]", ".")
  return new_str
```

```python
covid_train['Text'] = covid_train.apply(lambda row: replace_sentence(row),axis=1)
covid_validation['Text'] = covid_validation.apply(lambda row: replace_sentence(row),axis=1)
covid_test['Text'] = covid_test.apply(lambda row: replace_sentence(row),axis=1)
```

```python
from sklearn import model_selection, preprocessing


# label encode the target variable 
encoder = preprocessing.LabelEncoder()
y_train = encoder.fit_transform(covid_train['label'])
y_valid = encoder.fit_transform(covid_validation['label'])
y_test = encoder.fit_transform(covid_test['label'])
```

```python
class_names = encoder.classes_
print(class_names)
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
new_words=('=', 'et', 'al','e.g', 'sep', '2020', '2019', '1','2', '3', '0', '19', '4', '5', '6','7','8', '9', '10', ')', "also", "may", "ask")
for i in new_words:
    stopwords.append(i)
```

```python
import re
def tokenize(string):
    """
    Performs tokenization and string cleaning for the LitCovid dataset
    """
    string = re.sub(r"[^A-Za-z0-9(),!?\'`]", " ", string)
    string = re.sub(r"\s{2,}", " ", string)
    return string.lower().strip().split()
```

```python
from sklearn.feature_extraction.text import  TfidfVectorizer

# word level tf-idf
tfidf_vect = TfidfVectorizer(analyzer='word', stop_words=stopwords, tokenizer=tokenize, max_features=5000)
tfidf_vect.fit(covid_train['Text'])
X_train_tfidf =  tfidf_vect.transform(covid_train['Text'])
X_test_tfidf =  tfidf_vect.transform(covid_test['Text'])
```

# Logistic Regression Classifier

```python
from sklearn import linear_model

scikit_log_reg = linear_model.LogisticRegression(multi_class='ovr', solver='liblinear')
model=scikit_log_reg.fit(X_train_tfidf,y_train)
```

```python
from sklearn import metrics

def print_report(y_test,y_pred):

    report = metrics.classification_report(y_test, y_pred,
        target_names=class_names)
    print(report)
    print("accuracy: {:0.3f}".format(metrics.accuracy_score(y_test, y_pred)))
```

```python
y_pred = model.predict(X_test_tfidf)
print_report(y_test, y_pred)
```

```python
def get_predictions(predictions):
  y_pred =[]
  confidence=[]
  for i in range(0, len(predictions)):
        predicted_index = np.argmax(predictions[i])
        y_pred.append(predicted_index)
        confidence.append(predictions[predicted_index])
  return y_pred, confidence
```

# Captum

`transformers-interpret` is a wrapper around Captum, a library developed by the PyTorch team which implements several different explainability techniques, including saliency methods. Its visualization methods are a little easier to customize - particularly in displaying multiple records together. The code below uses LayerGradientXActivation -- a similar technique as the default `cls_explainer` (LIG) -- but can be extended with other attribution techniques or by performing attribution over different embeddings separately (word, position, etc.).

The following code runs on CPU to avoid CUDA memory errors.

```python
from captum.attr import LayerIntegratedGradients, LayerGradientXActivation
from captum.attr import visualization as viz
import torch
```

```python
def predict(inputs, attention_mask=None, model=model.model):
    return model(inputs, attention_mask=attention_mask)

def classifier_forward_func(inputs, attention_mask=None, class_idx=1):
  'Do a forward pass and return the probability'
  logits = predict(inputs=inputs, attention_mask=attention_mask)[0]
  probas = torch.softmax(logits, -1)
  return probas[:, class_idx]
```

```python
model.model.to('cpu')
lga = LayerGradientXActivation(classifier_forward_func, model.model.bert.embeddings)
```
