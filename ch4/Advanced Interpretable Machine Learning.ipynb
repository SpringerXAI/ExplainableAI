### Other Utilities

The following utility function will print performance at the threshold that achieves the highest F1-score and plot an interactive precision/recall/F1 curve.

```python
def proba_pred_detail(y_true, proba_pred, model_name=""):
  pr_df = pd.DataFrame(zip(*precision_recall_curve(y_true, probas_pred=proba_pred)),
                       columns=["Precision", "Recall", "Threshold"])
  pr_df['F1'] = 2 * pr_df['Precision'] * pr_df['Recall'] / (pr_df['Precision'] + pr_df['Recall'])

  fig = px.line(pr_df.melt("Threshold", var_name="Metric"), 
              "Threshold",
              y="value", 
              color="Metric", title=f"P/R/F1 Curve: {model_name}")
  fig.update_layout(template="plotly_white", hovermode="x")

  max_f1_thresh = pr_df['Threshold'].iloc[pr_df['F1'].values.argmax()]

  print(f"Performance at Max-F1 Threshold ({max_f1_thresh:.4}):")
  print(classification_report(y_true, proba_pred >= max_f1_thresh, digits=3))
  return fig  
```

Next, we would like to see the learned tree. The `pyoptree` library does not have built-in visualization, but we can use its internal components to build our own. The object returned by `OptimalTreeModel` and `OptimalHyperTreeModel` represents the tree using a row-based structure rather than a nested object. The code below recursively creates an directed graph, where each node corresponds to a node in the decision tree. This gives us a tree-like representation that preserves the concept of parent and child nodes.

```python
from graphviz import Digraph
```

```python
class Node(object):
    def __init__(self, 
                 left=None, 
                 right=None, 
                 is_leaf=False):
        self.left = left
        self.right = right
        self.is_leaf = is_leaf

def build_tree(current_depth,
                 current_row, 
                 max_depth,
                 model, 
                 feature_names):
    'Recursive function to construct a tree from the tabular format'
    
    this_node = Node()
    
    if current_depth >= max_depth:
        this_node.is_leaf = True
        this_node.title = f"Prediction: {np.array(model.c[current_row]).argmax()}"
        return this_node
    else:
        # get feature used for split by keying the table
        feature_idx = model.a[current_row].argmax()
        feature_name = feature_names[feature_idx]
        
        # get the split value (stored in scaled feature space)
        split_value_norm = float(model.b[current_row])
        
        # convert split value to original domain
        X_max, X_min = model.normalizer[feature_name]
        split_value = split_value_norm * (X_max - X_min) + X_min
        
        # get next rows
        left_row = current_row * 2
        right_row = current_row * 2 + 1
        
        this_node.title = f"{feature_name} < {round(split_value, 2)}"
        this_node.left = build_tree(current_depth+1, 
                                      current_row=left_row,
                                      max_depth=max_depth,
                                      model=model,
                                      feature_names=feature_names)
        this_node.right = build_tree(current_depth+1, 
                                       current_row=right_row,
                                       max_depth=max_depth,
                                       model=model,
                                       feature_names=feature_names)
        
        return this_node
```

```python
root = build_tree(current_depth=0, 
                  current_row=1, 
                  max_depth=op_tree.D, 
                  model=op_tree, 
                  feature_names=raw_feature_names)
```

Once we have a directed graph representation, we can un-nest the individual nodes and extract the information into `graphviz`. The function below simply creates a list containing each of the `Node` objects in the fitted tree. Each of these nodes contains information about parent/child connections in the tree, which we'll pass to graphviz.

```python
def extract_tree(this_node):
    global all_nodes
    
    all_nodes.append(this_node)
    
    if this_node.is_leaf:
        pass
    
    else:
        extract_tree(this_node.left)
        extract_tree(this_node.right)
        
all_nodes = []
extract_tree(root)
```

Here, we create the `graphviz.Digraph` object and loop through nodes. We add these nodes to the graph and add edges between parent and child for all non-leaf nodes. Once this is done, we can view our tree.

```python
tree_viz = Digraph('optimal_tree')
for i, node in enumerate(all_nodes):
    # add the node to the graph
    tree_viz.node(str(i), node.title)
    
    # add connection to children
    if not node.is_leaf:
        left_id = np.where(np.array(all_nodes) == node.left)[0].item()
        tree_viz.edge(str(i), str(left_id), label="True")
        
        right_id = np.where(np.array(all_nodes) == node.right)[0].item()
        tree_viz.edge(str(i), str(right_id), label="False")
        
tree_viz
```

```python
# create coefficient set and set the value of the offset parameter
coef_set = riskslim.CoefficientSet(variable_names = slim_data['variable_names'], lb = -max_coefficient, ub = max_coefficient, sign = 0)
coef_set.update_intercept_bounds(X = slim_data['X'], y = slim_data['Y'], max_offset = max_offset)
```

```python
constraints = {
    'L0_min': 0,
    'L0_max': max_L0_value,
    'coef_set':coef_set,
}

# major settings (see riskslim_ex_02_complete for full set of options)
settings = {
    # Problem Parameters
    'c0_value': c0_value,
    #
    # LCPA Settings
    'max_runtime': 30.0,                               # max runtime for LCPA
    'max_tolerance': np.finfo('float').eps,             # tolerance to stop LCPA (set to 0 to return provably optimal solution)
    'display_cplex_progress': True,                     # print CPLEX progress on screen
    'loss_computation': 'fast',                         # how to compute the loss function ('normal','fast','lookup')
    #
    # LCPA Improvements
    'round_flag': True,                                # round continuous solutions with SeqRd
    'polish_flag': True,                               # polish integer feasible solutions with DCD
    'chained_updates_flag': True,                      # use chained updates
    'add_cuts_at_heuristic_solutions': True,            # add cuts at integer feasible solutions found using polishing/rounding
    #
    # Initialization
    'initialization_flag': True,                       # use initialization procedure
    'init_max_runtime': 120.0,                         # max time to run CPA in initialization procedure
    'init_max_coefficient_gap': 0.49,
    #
    # CPLEX Solver Parameters
    'cplex_randomseed': 0,                              # random seed
    'cplex_mipemphasis': 0,                             # cplex MIP strategy
}
```

```python
#model info contains key results
pprint.pprint(slim_classifier_info)
```

One downside of this implementation is that it does not have a prediction method. We can use the formula in the printed output to create a vectorized scorer.

`Pr(Y = +1) = 1.0/(1.0 + exp(-(-2 + score))`, where `score` is each observation's score defined by the SLIM solution.

```python
def predict_slim_proba(data, coef):
  if data.shape[1] == (coef.shape[0] - 1):
    design_matrix = np.concatenate([np.ones((data.shape[0], 1)), 
                                    data], 
                                   axis=1)
  elif data.shape[1] == coef.shape[0]:
    design_matrix = data 
  else:
    raise ValueError("Data must have the same number of columns as coef or one fewer")
  
  scores = design_matrix @ coef 
  probas = 1 / (1 + np.exp(-(-2 + scores)))
  return probas
```

```python
slim_probs = predict_slim_proba(X_valid_binary_s, slim_classifier_info['solution'])

proba_pred_detail(y_valid_raw, slim_probs)
```
